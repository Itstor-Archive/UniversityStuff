```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
PI = math.pi

sampling_rate = 30 # 30Hz
ts = 1.0/sampling_rate # t=0s -> t=1s
t = np.arange(0, 1, ts)

signal = [0.00, 7.83, 7.09, 0.95, -1.83, -0.00, 0.59, -1.95, -3.09, -0.59, 1.73, 0.82, -0.95, -0.72, 0.24, 0.00, -0.24, 0.72, 0.95, -0.82, -1.73, 0.59, 3.09, 1.95, -0.59, -0.00, 1.83, -0.95, -7.09, -7.83]
```

# Plotting Sinyal Awal

```python
plt.figure(figsize = (8, 6))
plt.plot(t, signal, 'r')
plt.ylabel('Amplitude')

plt.show()
```

```python
def dft(seq):
    N = len(seq)
    y_real = [0] * N
    y_img = [0] * N

    for i in range(N):
        for j in range(N):
            e = -i * 2 * PI * j / N
            y_real[i] += seq[j] * math.cos(e)
            y_img[i] += seq[j] * math.sin(e)
    
    return y_real, y_img
```

```python
def inverse_dft(y_real, y_img):
    N = len(y_real)
    seq = [0] * N
    
    for i in range(N):
        for j in range(N):
            e = j * 2 * PI * i / N
            seq[i] += y_real[j] * math.cos(e) - y_img[j] * math.sin(e)
        
        seq[i] /= N
    
    return seq
```

# Soal 1

```python
y_real, y_img = dft(signal)

y_img = np.array(y_img)
y_real = np.array(y_real)
```

```python
N = y_real.size
n = np.arange(N)
T = N/sampling_rate
freq = n/T 

plt.figure(figsize = (8, 6))
plt.stem(freq, abs(y_img), 'b', markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('DFT Amplitude |X(freq)|')
plt.show()
```

# Soal 2

```python
y_real1 = y_real.copy()
y_img1 = y_img.copy()

for i in range(N):
    if i != 4:
        y_real1[i] = 0
        y_img1[i] = 0

inv1 = inverse_dft(y_real1, y_img1)
```

```python
plt.figure(figsize = (8, 6))
plt.plot(t, inv1, 'r')
plt.ylabel('Amplitude')

plt.show()
```
